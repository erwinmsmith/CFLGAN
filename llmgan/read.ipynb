```python
import json

try:
    with open('pubmed/train.json', 'r', encoding='utf-8') as file:
        data = json.load(file)
        
        # 打印前 10 个样本
        for i, sample in enumerate(data[:10]):
            print(f"Sample {i + 1}:")
            print(json.dumps(sample, indent=2, ensure_ascii=False))
            print("\n" + "-" * 50 + "\n")
            
except FileNotFoundError:
    print("文件未找到，请检查文件路径是否正确。")
except json.JSONDecodeError:
    print("文件格式不是有效的 JSON，请检查文件内容。")
```
